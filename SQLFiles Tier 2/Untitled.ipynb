```python
import sqlite3 as sql
import pandas as pd
```

```python
database = 'sqlite_db_pythonsqlite.db'
connection = sql.connect(database)
```

```python
query = '''
SELECT sub.months, COUNT(sub.memid) as monthyusage
FROM
    (SELECT memid, strftime('%m', starttime) as months
    FROM `Bookings`
    WHERE memid != 0) AS sub
GROUP BY sub.months
'''
```

```python
df = pd.read_sql_query(query, connection)
df
```
