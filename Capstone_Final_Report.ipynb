**Conclusion**

Accuracy tells you how many times the ML model was correct overall. Precision is how good the model is at predicting a specific category. Recall tells you how many times the model was able to detect a specific category. Since we care more about correctly identifying the fraudulent activity, we used recall as our ultimate metric.

Since our dataset is naturally imbalanced, it will make our accuracy score really high by default. That is why I chose to add a balanced weight class to increase recall rate. XGBoost and XGBoost/Random Forest had the highest accuracy score and F1 score, however Decision Tree and Random Forest has the highest recall which is the metric we are looking for. They both had a recall score of 81%. To decide between choosing Random Forest or Decision Tree as the best model, we look at the other measurements and Random Forest had higher accuracy and F1 score. Therefore, **Random Forest is our winning model** with an accuracy score of 90%, F1 score of 10%, and recall score of 81%.
