```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import datetime as dt
import os
import io
```

```python
# Data loading
data0 = pd.read_csv("a.us.txt")
data1 = pd.read_csv("aa.us.txt")
data2 = pd.read_csv("aaap.us.txt")
data3 = pd.read_csv("aaba.us.txt")
data4 = pd.read_csv("aac.us.txt")
data5 = pd.read_csv("aal.us.txt")
data6 = pd.read_csv("aamc.us.txt")
data7 = pd.read_csv("aame.us.txt")
data8 = pd.read_csv("aan.us.txt")
data9 = pd.read_csv("aaoi.us.txt")
# Data joining
df = [data0, data1, data2, data3, data4, data5, data6, data7, data8, data9]
key = ['a', 'aa', 'aaap', 'aaba', 'aac', 'aal', 'aamc', 'aame', 'aan', 'aaoi']
```

```python
name = ['Stock_Name', 'ID']
final = pd.concat(df, keys = key, names = name)
display(final)
```

```python
del final['OpenInt']
final.head()
```

```python
# There is no mode for this dataframe, therefore all numbers are unique values 
# Calculates the mean for all columns except for date
final.groupby(level=['Stock_Name'])[['Open', 'High', 'Low', 'Close', 'Volume']].mean()
```

```python
# Calculates the median for all columns except for date
final.groupby(level=['Stock_Name'])[['Open', 'High', 'Low', 'Close', 'Volume']].median()
```

```python
# Calculates the standard deviation for all columns except for date
final.groupby(level=['Stock_Name'])[['Open', 'High', 'Low', 'Close', 'Volume']].std()
```
